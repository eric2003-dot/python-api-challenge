```python
# Store Weather Py 

weather_df=pd.read_csv("Weather_Py.csv")
weather_df
weather_df.head()
```

```python
#humidity heatmap

locations = weather_df[["Lat", "Lng"]]
humidity = weather_df['Humidity'].astype(int)


# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
